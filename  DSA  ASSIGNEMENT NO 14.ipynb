Given a linked list ofÂ **N**Â nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**

```
Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.

```

**Example 2:**

</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detectAndRemoveLoop(head):
    if head is None or head.next is None:
        return head

    slow = head
    fast = head

    # Detect the loop
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    # If no loop found
    if fast is None or fast.next is None:
        return head

    # Reset slow pointer to head and move both pointers one step at a time
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Remove the loop
    fast.next = None

    return head

# Helper function to create a linked list with a loop
def createLinkedListWithLoop(arr, pos):
    if pos == -1:
        return createLinkedList(arr)

    head = Node(arr[0])
    current = head

    loopNode = None

    for i in range(1, len(arr)):
        node = Node(arr[i])
        current.next = node
        current = current.next

        if i == pos:
            loopNode = current

    if loopNode:
        current.next = loopNode

    return head

# Helper function to print the linked list
def printList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 1 -> 3 -> 4 (with a loop)
arr = [1, 3, 4]
head = createLinkedListWithLoop(arr, 1)

# Remove the loop from the linked list
modifiedHead = detectAndRemoveLoop(head)

# Print the modified linked list
printList(modifiedHead)
````

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**

```
Input:
LinkedList: 4->5->6
Output:457

```

**Example 2:**

</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def convertLinkedListToNumber(head):
    number = 0
    current = head

    while current:
        number = number * 10 + current.data
        current = current.next

    return number

# Helper function to create a linked list
def createLinkedList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    current = head

    for i in range(1, len(arr)):
        node = Node(arr[i])
        current.next = node
        current = current.next

    return head

# Example usage:
# Create the linked list: 4 -> 5 -> 6
arr = [4, 5, 6]
head = createLinkedList(arr)

# Convert the linked list to a number
number = convertLinkedListToNumber(head)

# Print the converted number
print(number)
````

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) aÂ **next**Â pointer to the next node,(ii) aÂ **bottom**Â pointerÂ to a linked list where this node is head.Each of theÂ sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order.Â **Note:**Â The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

```
Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output:Â 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```

**Example 2:**

</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def flattenLinkedList(head):
    if head is None or (head.next is None and head.bottom is None):
        return head

    nextList = flattenLinkedList(head.next)
    bottomList = flattenLinkedList(head.bottom)

    head.next = nextList

    if bottomList:
        current = nextList
        while current.next:
            current = current.next
        current.next = bottomList

    return head

# Helper function to create a multilevel linked list
def createMultilevelLinkedList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    current = head
    prev = None

    for i in range(1, len(arr)):
        if type(arr[i]) == int:
            node = Node(arr[i])
            current.next = node
            prev = current
            current = current.next
        else:
            node = Node(arr[i][0])
            current.bottom = node
            current = current.bottom
            for j in range(1, len(arr[i])):
                node.next = Node(arr[i][j])
                node = node.next

    return head

# Helper function to print the flattened linked list
def printFlattenedLinkedList(head):
    current = head
    while current:
        print(current.data, end="-> ")
        current = current.next
    print()

# Example usage:
# Create the multilevel linked list
arr = [
    5, 10, 19, 28,
    [7, 20, 22, 35],
    [8, [30, 45]],
    [50, [40]]
]
head = createMultilevelLinkedList(arr)

# Flatten the multilevel linked list
flattenedHead = flattenLinkedList(head)

# Print the flattened linked list
printFlattenedLinkedList(flattenedHead)
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def oddEvenLinkedList(head):
    if head is None or head.next is None:
        return head

    odd = head
    even = head.next
    oddHead = odd
    evenHead = even

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = evenHead
    if even:
        even.next = None

    return oddHead

# Helper function to create a linked list from an array
def createLinkedList(arr):
    if len(arr) == 0:
        return None

    head = ListNode(arr[0])
    current = head

    for i in range(1, len(arr)):
        node = ListNode(arr[i])
        current.next = node
        current = current.next

    return head

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 1 -> 2 -> 3 -> 4 -> 5
arr = [1, 2, 3, 4, 5]
head = createLinkedList(arr)

# Reorder the linked list
reorderedHead = oddEvenLinkedList(head)

# Print the reordered linked list
printLinkedList(reorderedHead)
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def rotateLinkedList(head, n, k):
    if head is None or k == 0:
        return head

    k %= n
    if k == 0:
        return head

    current = head
    prev = None

    # Find the last node and the previous node
    while current.next:
        prev = current
        current = current.next

    # Connect the last node to the head to form a circular list
    current.next = head

    # Traverse to the new head position
    for _ in range(n - k):
        prev = prev.next

    # Break the circular list
    newHead = prev.next
    prev.next = None

    return newHead

# Helper function to create a linked list from an array
def createLinkedList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    current = head

    for i in range(1, len(arr)):
        node = Node(arr[i])
        current.next = node
        current = current.next

    return head

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 2 -> 4 -> 7 -> 8 -> 9
arr = [2, 4, 7, 8, 9]
head = createLinkedList(arr)

# Rotate the linked list by k = 3 positions
rotatedHead = rotateLinkedList(head, 5, 3)

# Print the rotated linked list
printLinkedList(rotatedHead)
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def nextGreaterLinkedList(head):
    if head is None:
        return None

    current = head
    while current:
        greater = findNextGreater(current)
        current.data = greater if greater is not None else 0
        current = current.next

    return head

def findNextGreater(node):
    nextNode = node.next
    while nextNode:
        if nextNode.data > node.data:
            return nextNode.data
        nextNode = nextNode.next

    return None

# Helper function to create a linked list from an array
def createLinkedList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    current = head

    for i in range(1, len(arr)):
        node = Node(arr[i])
        current.next = node
        current = current.next

    return head

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 2 -> 7 -> 4 -> 3 -> 5
arr = [2, 7, 4, 3, 5]
head = createLinkedList(arr)

# Convert the linked list to next greater number format
convertedHead = nextGreaterLinkedList(head)

# Print the converted linked list
printLinkedList(convertedHead)
````

Given theÂ `head`Â of a linked list, we repeatedly delete consecutive sequences of nodes that sum toÂ `0`Â until there are no such sequences.

After doing so, return the head of the final linked list.Â  You may return any such answer.

(Note that in the examples below, all sequences are serializations ofÂ `ListNode`Â objects.)

**Example 1:**

```
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```

**Example 2:**

```
Input: head = [1,2,3,-3,4]
Output: [1,2,4]

```

**Example 3:**

</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeNegativeNodes(head):
    if head is None:
        return None

    # Check if the head node is negative
    while head and head.data < 0:
        head = head.next

    if head is None:
        return None

    prev = head
    current = head.next

    while current:
        if current.data < 0:
            prev.next = current.next
        else:
            prev = current

        current = current.next

    prev.next = None

    return head

# Helper function to create a linked list from an array
def createLinkedList(arr):
    if len(arr) == 0:
        return None

    head = Node(arr[0])
    current = head

    for i in range(1, len(arr)):
        node = Node(arr[i])
        current.next = node
        current = current.next

    return head

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Example usage:
# Create the linked list: 1 -> 2 -> 3 -> -3 -> 4
arr = [1, 2, 3, -3, 4]
head = createLinkedList(arr)

# Remove negative nodes from the linked list
modifiedHead = removeNegativeNodes(head)

# Print the modified linked list
printLinkedList(modifiedHead)
````
